# Linear regression between height and weight

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
wh_data = pd.read_csv("dataset/weight-height.csv")
```

```python
# Ploting distribution of heights for both gender
sns.set()
plt.figure(figsize=(8,8))
plt.hist(wh_data[wh_data["Gender"]=="Female"].Height, label="Female height distribution", alpha=0.5)
plt.hist(wh_data[wh_data["Gender"]=="Male"].Height, label="Male height distribution", alpha=0.5)
plt.legend()
```

```python
# Ploting distribution of weights for both gender
plt.figure(figsize=(8,8))
plt.hist(wh_data[wh_data["Gender"]=="Female"].Weight, label="Female weight distribution", alpha=0.5)
plt.hist(wh_data[wh_data["Gender"]=="Male"].Weight, label="Male weight distribution", alpha=0.5)
plt.legend()
plt.show()
wh_data.head()
```

```python
# Ploting data
plt.figure(figsize=(7, 7))
plt.scatter(wh_data[wh_data["Gender"]=="Female"].Height,
            wh_data[wh_data["Gender"]=="Female"].Weight, label="Female")
plt.scatter(wh_data[wh_data["Gender"]=="Male"].Height,
            wh_data[wh_data["Gender"]=="Male"].Weight, label="Male")
plt.legend(loc="upper right")
plt.show()
```

No difference between both gender.
